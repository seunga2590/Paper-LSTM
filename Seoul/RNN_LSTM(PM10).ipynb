```
def preprocess(all_files):
    data_0 = []  # SO2
    data_1 = []  # CO
    data_2 = []  # O3
    data_3 = []  # NO2
    data_4 = []  # PM10
    data_5 = []  # PM25
    data_6 = []  # tem
    data_7 = []  # rain
    data_8 = []  # hum
    data_9 = []  # pre
    data_10 = []  # wind_x
    data_11 = []  # wind_y 
    for f in all_files:
        with open(f, encoding='euc-kr') as c:
            csv_reader = csv.reader(c, delimiter=',')
            header = True
            for col in csv_reader:
                if header:
                    header = False
                    continue
                data_0.append(
                    float(col[1])) if col[1] != '' else data_0.append(0.0)
                data_1.append(
                    float(col[2])) if col[2] != '' else data_1.append(0.0)
                data_2.append(
                    float(col[3])) if col[3] != '' else data_2.append(0.0)
                data_3.append(
                    float(col[4])) if col[4] != '' else data_3.append(0.0)
                data_4.append(
                    float(col[5])) if col[5] != '' else data_4.append(0.0)
                data_5.append(
                    float(col[6])) if col[6] != '' else data_5.append(0.0)
                data_6.append(
                    float(col[7])) if col[7] != '' else data_6.append(0.0)
                data_7.append(
                    float(col[8])) if col[8] != '' else data_7.append(0.0)
                data_8.append(
                    float(col[9])) if col[9] != '' else data_8.append(0.0)
                data_9.append(
                    float(col[10])) if col[10] != '' else data_9.append(0.0)     
                data_10.append(
                    float(col[11])) if col[11] != '' else data_10.append(0.0)               

    data = np.zeros((len(data_0), 11))
    for i, d in enumerate(data):
        data[i, 0] = data_0[i]
        data[i, 1] = data_1[i]
        data[i, 2] = data_2[i]
        data[i, 3] = data_3[i]
        data[i, 4] = data_4[i]
        data[i, 5] = data_5[i]
        data[i, 6] = data_6[i]
        data[i, 7] = data_7[i]
        data[i, 8] = data_8[i]
        data[i, 9] = data_9[i]
        data[i, 10] = data_10[i]
    return data.astype(np.float32)
```

```
def make_dataset(data, seq_length=480, target_delay=24, strides=1,
                 mode='train', train_mean=None, train_std=None):

    assert mode in ['train', 'val', 'test']
    if mode is not 'train':
        if train_mean is None or train_std is None:
            print('Current mode is {}'.format(mode))
            print('This mode needs mean and std of train data')
            assert False

    # 정규화
    if mode is 'train':
        mean = train_data.mean(axis=0)   
        std = train_data.std(axis=0) 
    else: 
        mean = train_mean    
        std = train_std     
    data = (data - mean) / std
    
    # 입력, 타겟 데이터 생성
    sequence = []
    target = []
    for index in range(len(data) - seq_length - target_delay):
        if index % strides == 0:

            sequence.append(data[index:index+seq_length])
            target.append(data[index+seq_length+target_delay][4])

    if mode is 'train':
        return np.array(sequence), np.array(target), mean, std
    else:
        return np.array(sequence), np.array(target)
```

```
train_sequences, train_labels, train_mean, train_std = make_dataset(train_data, mode='train')
val_sequences, val_labels = make_dataset(val_data, mode='val', train_mean=train_mean, train_std=train_std)
test_sequences, test_labels = make_dataset(test_data, mode='test', train_mean=train_mean, train_std=train_std)
```

```
from tensorflow import keras

model = keras.Sequential()
model.add(keras.layers.LSTM(64, input_shape=train_sequences.shape[-2:], dropout=0.3, return_sequences=True))
model.add(keras.layers.LSTM(32, dropout=0.3))
model.add(keras.layers.Dense(2, activation='softmax'))
```

```
model.compile(loss='mean_squared_error', 
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              metrics=['mse'])
```

```
checkpoint_path = "/content/drive/My Drive/new/exp_rnn/cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-dropout-model.h5')
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)
```

```
# input에 tf.data.Dataset을 넣기 때문에 steps_per_epoch 지정해주어야 함.

history = model.fit(train_dataset, steps_per_epoch=100,
          validation_data = val_dataset, validation_steps=1, 
          epochs= 20,
          callbacks= [checkpoint_cb, early_stopping_cb])
```
