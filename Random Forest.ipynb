```python
import numpy as np

tmp = []
for each in df2['PM10']:
    if each <=30:
        tmp.append(0)
    elif each <=80:
        tmp.append(1)
    elif each <=150:
        tmp.append(2)
    elif each >=151:
        tmp.append(3)
    else:
        tmp.append(np.nan)
```

```python
from sklearn.metrics import mean_squared_error

rf_1 = RandomForestClassifier(bootstrap=True, criterion='gini',
                       max_depth=7, max_features=11, max_leaf_nodes=5,
                       min_samples_leaf=4, min_samples_split=61,n_estimators=1090,
                       n_jobs=-1,random_state=943, oob_score=True)
rf_1.fit(train_x,train_y.values.ravel())

print('Score on training set:{}' .format(rf_1.score(train_x,train_y)))
print('Score on test set:{}' .format (rf_1.score(test_x,test_y)))
mean_squared_error(test_y,rf_1.predict(test_x))
```

```python
print('Mean_Squared_Error = ', mean_squared_error(test_y, rf_1.predict(test_x)))
print('RMSE = ', mean_squared_error(test_y, rf_1.predict(test_x))**0.5)
```
